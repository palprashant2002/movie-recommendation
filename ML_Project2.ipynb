```python
#function to extract required data
def gen(arg):
    lst=[]
    arg=ast.literal_eval(arg)
    for x in arg:
        lst.append(x['name'])
    return lst
```

```python
#checking working of gen function
gen('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(gen)
```

```python
movies['keywords']=movies['keywords'].apply(gen)
```

```python
movies.iloc[0].keywords
```

```python
# to select few artist from whole cast
def scast(arg):
    lst=[] 
    arg=ast.literal_eval(arg)
    for x in arg:
        lst.append(x['name'])
        if len(lst)==5:
            break;
    return lst
```

```python
movies['cast']=movies['cast'].apply(scast)
```

```python
movies.iloc[0].cast
```

```python
movies.head(4)
```

```python
#function to fetch director name
def direc_cast(arg3):
    lst=[]
    arg=ast.literal_eval(arg3)
    for x in arg:
        if(x['job']=='Director'):
            lst.append(x['name'])
    return lst
```

```python
movies['crew']=movies['crew'].apply(direc_cast)
```

```python
movies.iloc[0].crew
```

```python
movies.head(3)
```

```python
#removing spaces between the names of genre,actor,director etc...
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#creating a tag column 
movies['sch']=movies['genres']+movies['crew']+movies['cast']+movies['keywords']
```

```python
movies.iloc[0].sch
```

```python
#creating new dataframe
up_data=movies[['movie_id','title','sch']]
```

```python
up_data.head()
```

```python
#converting the tag list into string
up_data['sch']=up_data['sch'].apply(lambda x:" ".join(x))
```

```python
 up_data.head()
```

```python
#converting all uppercase letters into lowercase 
up_data['sch']=up_data['sch'].apply(lambda x: x.lower())
```

```python
up_data.head()
```

```python
#creating an object to perform stemming
ps=PorterStemmer()
```

```python
#function to perform stemming
def pstem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return(" ".join(y)) 
```

```python
#stemming
up_data['sch']=up_data['sch'].apply(pstem)
```

```python
# Creating an object to perform Vectorization 
cv=CountVectorizer(max_features=5000,stop_words='english')  #only 5000 most occuring words are considered & stop words are removed 
```

```python
#performing vectorization
vector=cv.fit_transform(up_data['sch']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#finding similarity between movies
similarity=cosine_similarity(vector)
```

```python
#function to pic out movies
def recommend(inp_movie):
    mov_ind=up_data[up_data['title']==inp_movie].index[0]
    dist=similarity[mov_ind]
    mov_lst=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:10]
    for i in mov_lst:
        print(up_data.iloc[i[0]].title)
    print(type(dist))
    print(dist.shape)
```
